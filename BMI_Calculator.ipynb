```python
print("BMI CALCULATOR")
```

```python
weight=float(input("Enter weight in kg:"))
height=float(input("Enter height in meters:"))
             
```

```python
bmi=weight/(height*height)
print("Your BMIis:",bmi)
```

```python
if bmi<18.5:
    print("Underweight")
elif bmi<24.9:
    print("Normalweight")
elif bmi<29.9:
    print("Overweight")
else:
    print("Obese")
```
